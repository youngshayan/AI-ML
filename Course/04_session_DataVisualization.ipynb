# Distribution

```python
%pip install scipy
```

```python
from numpy import arange
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_space = arange(-5, 5, 0.001)
mean = 0.0
stdev = 1.0

pdf = norm.pdf(sample_space, mean, stdev)

cdf = norm.cdf(sample_space, mean, stdev)

fig, ax = plt.subplots(1,2, figsize=(8,4))
fig.suptitle('normal distribution')
ax[0].plot(sample_space, pdf, color = 'purple')
ax[0].set_title('pdf')
ax[1].plot(sample_space, cdf, color = 'cyan')
ax[1].set_title('cdf');
```

```python
from scipy.stats import gamma

sample_space = arange(0, 20, 0.001)
alpha = 5
beta = 1
scale = 1/beta

pdf = gamma.pdf(sample_space, alpha, scale)


cdf = gamma.cdf(sample_space, alpha, scale)

fig, ax = plt.subplots(1,2, figsize=(8,4))
fig.suptitle('gamma distribution')
ax[0].plot(sample_space, pdf, color = 'purple')
ax[0].set_title('pdf')
ax[1].plot(sample_space, cdf, color = 'cyan')
ax[1].set_title('cdf');
```

```python
from scipy.stats import uniform

sample_space = arange(-5, 5, 0.001)
scale = 8
loc = -4

pdf = uniform.pdf(sample_space, loc, scale)

cdf = uniform.cdf(sample_space, loc, scale)

fig, ax = plt.subplots(1,2, figsize=(8,4))
fig.suptitle('uniform distribution')
ax[0].plot(sample_space, pdf, color = 'purple')
ax[0].set_title('pdf')
ax[1].plot(sample_space, cdf, color = 'cyan')
ax[1].set_title('cdf');
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

x = np.array([1,2,3,4,5])
y = x * 2

ax.plot(x + 2 ,y,marker = 'x')                      # blue
ax.plot(x + 1 ,y,marker = 'o')                      # orange
ax.plot(x ,y ,marker = 'p')                         # green
ax.plot(x - 1 ,y,marker = '*', markersize = 5)      # red
ax.plot(x - 2 ,y,marker = '1', markersize = 10)     # purple
ax.plot(x - 3 ,y,marker = 'v', markersize = 15)     # brown
ax.plot(x - 4 ,y,marker = 'h', markersize = 25);    # pink
```

## Regplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme()

fig , ax = plt.subplots()

price = [921, 3087, 7196, 4201, 982, 3534, 9930, 8518, 12825, 7404, 4032, 878, 1240, 945, 10446, 5530, 720, 2846, 3064]
carat = [0.42, 0.76, 1.21, 0.91, 0.4, 1.02, 1.24, 1.5, 1.31, 1.13, 0.57, 0.41, 0.49, 0.42, 1.14, 0.91, 0.32, 0.77, 0.8]
df = pd.DataFrame()
df["price"] = price
df["carat"] = carat

sns.regplot(data= df,x= df["carat"],y=df["price"], ax= ax)
ax.set_title("carat vs price for diamonds")
ax.set_xlabel("carat")
ax.set_ylabel("price");
```
